```python
import geopandas as gpd
import folium
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = gpd.read_file("Population_District_Level.shp")
df
```

```python
df['toc_do'] = df['Pop_2019'] / df['Pop_2017']
district = df.sort_values(['toc_do'], axis = 0, ascending=False)
data_top10 = district[:10]
data = district
data 
```

```python
response = gpd.read_file('response3.json')
response
#response.head()
```

```python
data_json = response.to_crs(epsg=32648)
data_json
```

```python
new_data = gpd.sjoin(data_json, data[0:10])
new_data
```

```python
District_name = list(set(new_data['Dist_Name']))
District_name
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np
```

```python
from sklearn.cluster import KMeans
i = 0
result = []
for i in range(len(District_name)):
  x_  = new_data[new_data.Dist_Name == District_name[i]]
  kmeans = KMeans(n_clusters=20)
  a = pd.Series(x_['geometry'].apply(lambda p: p.x))
  b = pd.Series(x_['geometry'].apply(lambda p: p.y))
  X = np.column_stack((a,b))
  y_kmeans = kmeans.fit_predict(X)
  y_kmeans = kmeans.fit_predict(X)
  x_['cluster'] = y_kmeans
  result.append([District_name[i],x_])
result
```

```python
max_cluster = []
for i in result:
  temp1 = i[1]['cluster'].value_counts()
  print(temp1)
  t = int(input())
  max_cluster.append([i[0], t])
  
```

```python
print(max_cluster)
```

```python
filter_district = []
i = 0
for ele in max_cluster:
  temp_data = result[i][1][result[i][1].cluster == ele[1]]
  filter_district.append([ele[0], temp_data])
  i+=1
filter_district
```

```python
toa_do = []
for ele in filter_district:
  toa_do.append(ele[1])
toa_do
```
